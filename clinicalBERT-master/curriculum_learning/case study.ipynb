```python
import pandas as pd

# 取消省略
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)

# 读文件
path = '/mnt/nvme2/yyc/clinicalbert/data/discharge/train_with_difficulty_kmeans+distance_discharge.csv'
df = pd.read_csv(path)

# 按难度分组，每组输出前 10 个样本，并在每个样本后加分隔线
for diff in ['easy', 'middle', 'hard']:
    print(f"\n===== Difficulty: {diff.upper()} =====\n")
    texts = df[df['difficulty'] == diff]['TEXT'].head(5)
    for i, text in enumerate(texts, 1):
        print(f"--- Sample {i} ---\n")
        print(text, "\n")
        print("-" * 80)   # 80 个 “-” 作为分隔符

```

```python
import pandas as pd

# 取消省略设置
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)

# 读文件
path = '/mnt/nvme2/yyc/clinicalbert/data/3days/train_with_difficulty_kmeans+distance_3days.csv'
df = pd.read_csv(path)

# 打印前10行，每个样本及其相同 ID 的所有样本，列单独分行，样本组之间用分隔线
for i, (_, row) in enumerate(df.head(5).iterrows(), 1):
    this_id = row['ID']
    group = df[df['ID'] == this_id]
    print(f"=== Group for ID = {this_id} (from sample #{i}) ===")
    for j, (_, r) in enumerate(group.iterrows(), 1):
        print(f"--- Sub-sample {j} of ID {this_id} ---")
        for col in df.columns:
            print(f"{col}: {r[col]}")
        print()  # 空行分隔子样本
    print("=" * 80)  # 分隔不同 ID 组


```
